```python
loan_data = pd.read_csv('loan_data (1).csv')
loan_df = pd.DataFrame(loan_data)
loan_df
```

```python
l_df = loan_df.dropna()
l_df
```

```python
defaulters = l_df.loc[l_df['TARGET'] == 1]
non_defaulters = l_df.loc[l_df['TARGET'] == 0]
default_subset = defaulters 
default_subset2 = default_subset
default_subset3 = default_subset2
default_subset4 = default_subset3
default_subset5 = default_subset4
default_subset6 = default_subset5
default_subset7 = default_subset6
default_subset8 = default_subset7
default_subset9 = default_subset8
default_subset10 = default_subset9
default_subset11 = default_subset10
default_subset12 = default_subset11
default_subset13 = default_subset12
default_subset14 = default_subset13
default_subset15 = default_subset14
```

```python
duplicated_defaulters = pd.concat([defaulters, default_subset, default_subset2, default_subset3, default_subset4, default_subset5, default_subset6, default_subset7, default_subset8, default_subset9, default_subset10, default_subset11, default_subset12, default_subset13, default_subset14])
```

```python
balanced_data = pd.concat([duplicated_defaulters, non_defaulters])
```

```python
counts = balanced_data['TARGET'].value_counts()
counts_df = pd.DataFrame(counts)
c = counts_df['TARGET']
c
```

```python
defaulters = c[1]
non_defaulters = c[0]
total_payers = c[1] + c[0]
total_payers
```

```python
default_percentage = round((defaulters / total_payers)*100, 1)
non_default_percentage = round((non_defaulters / total_payers)*100, 1)
print(default_percentage)
print(non_default_percentage)
```
